```python
# Defining the cost(error) function
def cost(x,y,m):
    total_cost = 0
    M = x.shape[0]
    for i in range(M):
        X = x[i]
        Y = y[i]
        total_cost += (1/M)*(Y - (m*X).sum())**2
    return total_cost.mean()
```

```python
# Helper Function to apply gradient descent 
def step_gradient(x,y,learning_rate,m):
    m_slope = np.zeros(x.shape[1])
    M = x.shape[0]
    for i in range(M):
        X = x[i]
        Y = y[i]
        for j in range(x.shape[1]):
            m_slope[j] += (-2/M)*(Y - (m*X).sum())*X[j]
        
        for j in range(x.shape[1]):
            m[j] -= learning_rate*m_slope[j]
    
    return m
```

```python
# Function to apply gradient descent on given data and printing cost for each iteration
def gd(x,y,learning_rate,num_iterations):
    m = np.zeros(x.shape[1])
    for i in range(num_iterations):
        m = step_gradient(x,y,learning_rate,m)
        print(i,cost(x,y,m))
    return m
```

```python
# Initialising learning rate and number of iterations
def run(x,y):
    learning_rate = 0.0005
    num_iterations = 2000
    m = gd(x,y,learning_rate,num_iterations)
    return m
```

```python
m = run(x,y)
```

```python
# Predicting values of test data using values of m 
M = test.shape[0]
Y_pred = np.zeros(M)
for i in range(M):
    Y_pred[i] = (m*test[i]).sum()
```
